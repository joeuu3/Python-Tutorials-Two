```python
# Generating x and y points to plot
x1 = np.arange(0,10*(np.pi),0.1)
y1 = np.cos(x1) # cosine function from Numpy library

# potting the points
plt.plot(x1,y1, label = 'cosine')

plt.legend()   # shows the legend with labels
plt.xlabel('Angle')   # naming the x axis
plt.ylabel('Amplitude')  # naming the y axis

# Generating x and y points for the second plot
x2 = np.arange(0,10*(np.pi),0.1)
# decaying cosine function
y2 = np.cos(x2)*0.1*np.arange(10*(np.pi),0,-0.1)
# potting the points
plt.plot (x2,y2,label = 'Decaying cosine')
# show a lengend on the plot
plt.legend()
# gives a title to the plot
plt.title('Two functions on the same graph')
# shows the graph, and removes the text output.
plt.show()
```

## LET US PLOT AVG. AREA INCOME AGAINST PRICE FROM US_HOUSING DATASET.

```python
# Plotting Avg. Area Income vs Price
plt.plot(US_House['Avg. Area Income'], US_House['Price'], color='blue', marker='o')

# Giving title and label names
plt.title('Avg. Area Income Vs Price',fontsize=18)
plt.xlabel('Avg. Area Income', fontsize=18)
plt.ylabel('Price',fontsize=18)
plt.grid(True)
plt.show()
```

```python
# Adjusting the figure size using rcParams function
plt.rcParams['figure.figsize'] = [12,8]

# Scatter plot
plt.scatter(US_House['Avg. Area Income'],US_House['Price'], color='red',marker='o')

# Giving title and label
plt.title('Avg. Area Income Vs Price', fontsize=14)
plt.title('Avg. Area Income', fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.grid(True)
plt.show()
```
